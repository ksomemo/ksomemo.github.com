```python
df = pd.DataFrame({"n": range(10)})
df = df.assign(
    sp_z=sp.stats.zscore(df["n"]),
    ddof_t_z=(df["n"] - df["n"].mean()) / df["n"].std(),
    ddof_f_z=(df["n"] - df["n"].mean()) / df["n"].std(ddof=False)
)
ddof=False

print(df.min())
print(df.max())

df.plot(subplots=True, figsize=(10, 4), layout=(1, 4))
plt.tight_layout()
```
